# What is Data Visualization
- Data visualization is the graphical representation of information and data.
- Visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers and patterns iin data.

## What is Matplotlib ?
- Matplotlib is a plotting library for python programming language and its numerical mathematics extension NumPy.
- Matplotlib is a python library used for Data Visualizatation.
- Matplotlib is 2D and 3D plotting library.
- It was Iintoduced by john hunter in year 2002.

```python
import matplotlib.pyplot as plt                           # Importing matplotlib with alias plt

# Creating Data for visualization

x = [1,2,3,4]                                             # Creating x variable 
y = [2,4,6,8]                                             # Creating y variable

plt.plot(x,y)                                             # Simple plotting of x and y

plt.show()                                                # Showing plot
```

```python
# Bar graph

plt.bar(x,y)                                              # Plotting bar graph on variables x and y

plt.show()                                                # Showing graph
```

```python
# Bar graph

c = ['r','y','g','b']                                     # creating a list of colors to add to bars

plt.bar(x,y,color = c)                                    # adding color parameter to graph

plt.show()                                                # showing plot
```
