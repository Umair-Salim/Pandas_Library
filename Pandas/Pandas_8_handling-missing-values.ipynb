# dropna & fillna

```python
import pandas as pd                                                      # Importing pandas

var = pd.read_csv('test1.csv')                                           # reading a dataframe into variable "var"

var                                                                      # Printing var
```

```python
var.dropna()                                                               # Dropping "All" NaN values
```

```python
var.dropna(axis= 0)                                                        # Dropping only "Rows" with NaN values
```

```python
var.dropna(how = 'all')                                                  # Dropping NaN values using "how" parameter
```

```python
var.dropna(subset = ['points'])                                          # Dropping NaN values of perticular column using "subset" parameter
```

```python
var.dropna(inplace = True)                                            # Using "Inplace" parameter to save changes of "dropna" method 

var
```

```python
var.dropna(thresh = 1)                                             # Dropping only rows wtih 1 NaN value
```

```python
var.fillna('python')                                                # Filling all NaN values with "python"
```

```python
var.fillna({'points':'python', 'score':'c', 'Age': 'java'})         # Creating a dictionary to fill each column's NaN value
                                                                    # with specific value
```

```python
var.fillna(method = 'ffill')                                       # filling NaN values with the one before them along rows
```

```python
var.fillna(method = 'bfill', axis = 1)                            # filling NaN values with one after them,but along columns
```

```python
var.fillna(12,inplace = True)                                       # Using "inplace" parameter with same function as in "dropna"

var
```

```python
var.fillna('python', limit = 2)                                    # filling NaN values with python but only first two in each columns
```
