#  Merge

```python
import pandas as pd                                             # Importing pandas

var1 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})          # Creating dataframe var1
var2 = pd.DataFrame({'A':[1,2,3,4],'c':[21,22,23,24]})          # Creating dataframe var2

pd.merge(var1,var2,on='A')                                      # Merging var1 & var2 on the common column A
```

```python
pd.merge(var2,var1,on='A')                                     # Changing the order of columns after merging
```

```python
var3 = pd.DataFrame({'A':[1,2,3,5],'d':[21,22,23,24]})       # Creating a new dataframe var3, with one different value in col A
                                                             # Only the common values of the column which is passed in parameter
                                                             # "on" will be merged, rest will be dropped

pd.merge(var1,var3,on='A')
```

```python
pd.merge(var1,var3, how = 'left')                          # Merging with left column using "how" parameter
```

```python
pd.merge(var1,var3, how = 'right')                         # Merging with right column using "how" parameter
```

```python
pd.merge(var1,var3, how = 'inner')                        # Merging with only common values column using "how" parameter
```

```python
pd.merge(var1,var3, how = 'outer')                       # Merging with all values column using "how" parameter
```

```python
pd.merge(var1,var3, how = 'outer', indicator = True)    # Merging with all values column using "how" parameter,
                                                        # also using "indicator" parameter 
```

```python
pd.merge(var1,var3,left_index= True, right_index= True)# Merging by defining indexes
```

```python
pd.merge(var1,var3,left_index= True, right_index= True,suffixes = ('name','python')) # Changing suffixes
```

## Concat

```python
d1 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]}) # Creating dataframe d1
d2 = pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]}) # Creating dataframe d2

pd.concat([d1,d2])                                   # Concating dataframes
```

```python
pd.concat([d1,d2], axis = 1)                      # concating along the column axis
```

```python
d3 = pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]}) 
d4 = pd.DataFrame({'A':[1,2],'B':[21,22]}) 

pd.concat([d3,d4], axis = 1)                   # concating unequal dataframes
```

```python
pd.concat([d3,d4], axis = 1, join = 'inner')  # Concating oly common values
```

```python
pd.concat([d1,d2], axis = 0, keys = ["d1","d2"]) # Displaying keys along rows
```

```python
pd.concat([d1,d2], axis = 1, keys = ["d1","d2"]) #displaying keys along columns
```
