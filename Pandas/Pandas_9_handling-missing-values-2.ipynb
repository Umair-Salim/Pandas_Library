#  Replace

```python
import pandas as pd                                             # Importing pandas

var = pd.read_csv('test1.csv')                                  # Reading dataframe into var

var                                                             # Printing var
```

```python
var.replace(to_replace=24.0, value='python')                        # Replacing a specific value with "python"
```

```python
var.replace([20.0,24.0,30.0,23.0], 'python')                      # Replacing a list of values with "python"
```

```python
var.replace("[A-Za-z]","python", regex = True)                    # Replacing values using "regex" parameter
```

```python
var.replace({'First Name':'[A-Z]'},22,regex=True)                  # Replpacing values using "regex" parameter, but in only
                                                                   # one column
```

```python
var.replace(24.0, method = 'ffill')                               # Replacing values using forward method
```

```python
var.replace(24.0, method = 'bfill')                              # Replacing values using backward method
```

```python
var.replace(24.0,method = 'ffill', limit = 1)                     # replacing values and setting limit to 1
```

## Interpolate

```python
var.interpolate()                                                     # Interpolating values
```

```python
var.interpolate(method = 'linear')                                    # Interpolating values using linear method
```

```python
var.interpolate(method = 'linear', axis = 0)                          # Interpolating values along rows using linear method
```

```python
var.interpolate(limit_direction = 'backward', limit = 1)            # Setting limit and direction of limit while interpolating
```
