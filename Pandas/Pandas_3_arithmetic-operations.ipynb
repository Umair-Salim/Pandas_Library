# Arithmetic Operations

```python
import pandas as pd                                                             #Importing Pandas 

var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})                               #Creating Dataframe named var

var                                                                             #printing dataframe
```

Pandas is a library which is mostly used to perform data analysis on structured data. One of the many benefits of this library is that it can perform "Arithmetic" and "Logical" operations on dataframes. In this notebook we will explore some basic arithmetic and logical operations which can be done on columns of a dataframe using this popular library.

```python
# Addition

var["C"] = var["A"] + var["B"]                                                  #adding col A & B, Printing result in C                                               

var                                                                             #Printing dataframe
```

```python
# Subtraction

var["D"] = var["A"] - var["B"]                                                 #Subtracting Col A & B, Printing result in D

var                                                                            #Printing dataframe
```

```python
# Multiplication

var["E"] = var["A"] * var["B"]                                                 #Multplying Col A & B, Printing result in E

var                                                                            #Printing dataframe
```

```python
# Division

var["F"] = var["A"] / var["B"]                                                #Dividing Col A & B, Printing result in F         

var                                                                           #Printing dataframe
```

## Logical operations

```python
var1 = pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})                   #Creating new dataframe called var1

var1                                                                         #Printing dataframe
```

```python
# Logical Operators

var1["Python"] = var1["A"] <= 20                                             #Creating new Col and printing Logical result of
                                                                             #Statement i.e Print where Col A is less then or 
                                                                             #equal to 20

var1["Python_1"] = var1["B"] >= 16                                           #Creating new Col and printing Logical result of
                                                                             #Statement i.e Print where Col B is greater then  
                                                                             #or equal to 16                                     

var1                                                                         #Printing dataframe
```
