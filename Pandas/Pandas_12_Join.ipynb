#  Join

```python
import pandas as pd

var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"C":[10,20,30,40],"D":[11,22,33,44]})

var1.join(var2)
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"C":[10,20],"D":[11,22]})

var3.join(var4)
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index = ['a','b','c','d'])
var4 = pd.DataFrame({"C":[10,20],"D":[11,22]}, index = ['a','b'])

var3.join(var4)
```

```python
var4.join(var3)
```

```python
var4.join(var3, how = 'left')
```

```python
var4.join(var3, how = 'right')
```

```python
var4.join(var3, how = 'outer')
```

```python
var4.join(var3, how = 'inner')
```
