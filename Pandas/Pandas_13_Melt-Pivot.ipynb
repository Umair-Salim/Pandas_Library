# Melt

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})

var
```

```python
pd.melt(var)
```

```python
pd.melt(var, id_vars = ['days'])
```

```python
pd.melt(var, id_vars = ['days'],var_name = 'python')
```

```python
pd.melt(var, id_vars = ['days'],var_name='python',value_name='pak')
```

# Pivot

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6],
                    "st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})

var
```

```python
var.pivot(index='days', columns='st_name')
```

```python
var.pivot(index='days', columns='st_name', values='eng')
```

```python
var1 = pd.DataFrame({"days":[1,1,1,1,2,2],
                    "st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})

var1.pivot_table(index='st_name', columns='days',aggfunc='mean')
```

```python
var1.pivot_table(index='st_name', columns='days',aggfunc='sum')
```

```python
var1.pivot_table(index='st_name', columns='days',aggfunc='sum', margins='True')
```
