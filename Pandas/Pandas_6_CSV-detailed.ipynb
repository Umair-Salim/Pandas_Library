# Read CSV Files
In this notebook we will explore different ways of reading a csv file while using different parameters and how can we manipulate it while reading.Some of the parameters used are:
- `nrows=`
- `usecols=[""]`
- `skiprows=`
- `index_col=`
- `header=`
- `names=[""]`
- `header=None,prefix=`

```python
import pandas as pd                                                            #Importing Pandas

csv_1 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv")           #Reading CSV file

csv_1                                                                          #printing CSV file
```

```python
csv_2 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",nrows=3)           #Reading CSV file with only top 3 rows

csv_2                                                                                  #priinting CSV file
```

```python
csv_3 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",usecols=["Period","UNITS"]) #Reading CSV file with only 2 colmns

csv_3                                                                                           #printing CSV file
```

```python
csv_4 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",usecols=[1,5])   #Reading CSV file with only 2 cols using index

csv_4                                                                                #printing CSV file 
```

```python
csv_4 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",skiprows=[0,3]) #Reading CSV file while skipping rows

csv_4                                                                               #printing CSV file
```

```python
csv_5 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",index_col="Period") #Reading CSV file while setting period col as index

csv_5                                                                                   #printing CSV file
```

```python
csv_6 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",header=3) #Reading CSV file while setting row 3 as header

csv_6                                                                         #printing CSV file
```

```python
csv_7 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",names=["col1","col2","col3","col4",
                                                                           "col5","col6","col7","col8",
                                                                           "col9","col10","col11","col12",
                                                                           "col13","col14"]) #Reading CSV file while setting
                                                                                             #Header names using names parameter

csv_7                                                                                        #printing CSV file
```

```python
csv_8 = pd.read_csv("C:\\Users\\92333\\Documents\\Pandas\\test.csv",header=None, prefix = 'Col') #reading CSV file while changing header

csv_8                                                                                            #PRinting CSV file
```
